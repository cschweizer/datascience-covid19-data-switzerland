# Cases all Cantons

```python
import pandas as pd
import seaborn as sn
import plotly as py
import plotly.express as px
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import io
import requests
import glob
import os

from plotly.subplots import make_subplots

cantons = ['AG', 'AI', 'AR', 'BE', 'BL', 'BS', 'FR', 'GE', 'GL', 'GR', 'JU', 'LU', 'NE', 'NW', 'OW', 'SG', 'SH', 'SO', 'SZ', 'TG', 'TI', 'UR', 'VD', 'VS', 'ZG', 'ZH']

# Get the latest File
url = 'https://raw.githubusercontent.com/cschweizer/datascience-covid19-data-switzerland/master/CovidData.csv'
s=requests.get(url).content
df = pd.read_csv(io.StringIO(s.decode('utf-8')))
df['date'] = pd.to_datetime(df['date'])

# Verlauf pro Kanton
for canton in cantons:
    dfCanton = df[df['canton'] == canton]
    dfCanton['seven_day_conf'] = dfCanton['daily_conf'].rolling(window=7).mean()
    dfCanton['seven_day_dec'] = dfCanton['daily_deceased'].rolling(window=7).mean()    
    # Create figure with secondary y-axis
    fig = make_subplots(specs=[[{"secondary_y": True}]])

    fig.add_trace(
        go.Scatter(
        x = dfCanton['date'],
        y = dfCanton['seven_day_conf'],
        mode = 'lines',
        marker = dict(color = '#00008B'),
        name = 'Anzahl Positive'
    ),
        secondary_y=False,
    )

    fig.add_trace(
        go.Scatter(
        x = dfCanton['date'],
        y = dfCanton['seven_day_dec'],
        mode = 'lines',
        marker = dict(color = '#4682B4'),
        name = 'Todesf채lle'
    ),
        secondary_y=True,
    )

    # Add figure title
    fig.update_layout(
        title_text= canton + ": Ansteckungen und Todesf채lle"
    )

    # Set x-axis title
    fig.update_xaxes(title_text="Datum")

    # Set y-axes titles
    fig.update_yaxes(title_text="<b>Anzahl Positiv</b>", secondary_y=False)
    fig.update_yaxes(title_text="<b>Todesf채lle</b>째", secondary_y=True)

    fig.show()
```
