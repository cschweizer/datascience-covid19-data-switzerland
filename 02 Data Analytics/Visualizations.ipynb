# Import and Transform Data

```python
import pandas as pd

df = pd.read_csv('../Data/CovidData.csv')
df['date'] = pd.to_datetime(df['date'])

df.head()
```

# Cases vs Hospitalizations: Switzerland

```python
import seaborn as sn
import plotly as py
import plotly.express as px
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import glob
import os
from plotly.subplots import make_subplots

dfSwitzerland = df[['daily_conf', 'new_hosp']].groupby(df['date']).sum().rolling(window=7).mean().reset_index()

# dfSwitzerland = df.groupby(['date', as_index=False)]).agg('sum')

dfSwitzerland

# Create figure with secondary y-axis
fig = make_subplots(specs=[[{"secondary_y": True}]])

fig.add_trace(
    go.Scatter(
    x = dfSwitzerland['date'],
    y = dfSwitzerland['daily_conf'],
    mode = 'lines',
    marker = dict(color = '#00008B'),
    name = 'Anzahl Positive'
),
    secondary_y=False,
)

fig.add_trace(
    go.Scatter(
    x = dfSwitzerland['date'],
    y = dfSwitzerland['new_hosp'],
    mode = 'lines',
    marker = dict(color = '#4682B4'),
    name = 'Hospitalisierungen'
),
    secondary_y=True,
)

# Add figure title
fig.update_layout(
    title_text=  "Ansteckungen und Hospitalisierungen"
)

# Set x-axis title
fig.update_xaxes(title_text="Datum")

# Set y-axes titles
fig.update_yaxes(title_text="<b>Anzahl Positive (7-day mean)</b>", secondary_y=False)
fig.update_yaxes(title_text="<b>Hospitalisierungen (7-day mean)</b>°", secondary_y=True)

fig.show()
```

# Cases vs hospitalizations: each Cantons

```python
allCantons = ['AG', 'AI', 'AR', 'BE', 'BL', 'BS', 'FR', 'GE', 'GL', 'GR', 'JU', 'LU', 'NE', 'NW', 'OW', 'SG', 'SH', 'SO', 'SZ', 'TG', 'TI', 'UR', 'VD', 'VS', 'ZG', 'ZH']
#cantons = ['ZH']

# Verlauf pro Kanton
for canton in allCantons:
    dfCanton = df[df['canton'] == canton]
    dfCanton['seven_day_conf'] = dfCanton['daily_conf'].rolling(window=7).mean()
    dfCanton['seven_day_new_hosp'] = dfCanton['new_hosp'].rolling(window=7).mean()    
    # Create figure with secondary y-axis
    fig = make_subplots(specs=[[{"secondary_y": True}]])

    fig.add_trace(
        go.Scatter(
        x = dfCanton['date'],
        y = dfCanton['seven_day_conf'],
        mode = 'lines',
        marker = dict(color = '#00008B'),
        name = 'Anzahl Positive'
    ),
        secondary_y=False,
    )

    fig.add_trace(
        go.Scatter(
        x = dfCanton['date'],
        y = dfCanton['seven_day_new_hosp'],
        mode = 'lines',
        marker = dict(color = '#4682B4'),
        name = 'Hospitalisierungen'
    ),
        secondary_y=True,
    )

    # Add figure title
    fig.update_layout(
        title_text= canton + ": Ansteckungen und Hospitalisationen"
    )

    # Set x-axis title
    fig.update_xaxes(title_text="Datum")

    # Set y-axes titles
    fig.update_yaxes(title_text="<b>Anzahl Positiv (7-day mean)</b>", secondary_y=False)
    fig.update_yaxes(title_text="<b>Hospitalisierungen (7-day mean)</b>°", secondary_y=True)

    fig.show()
```
